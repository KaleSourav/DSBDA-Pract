## 8. Boxplot — Age by Sex

```python
sns.boxplot(x='sex', y='age', data=df)
plt.title('Age Distribution by Sex')
plt.show()
```

## 9. Boxplot — Age by Sex (Hue: Survived)

```python
sns.boxplot(x='sex', y='age', data=df, hue='survived')
plt.title('Age Distribution by Sex and Survival')
plt.show()
```

## 10. Stripplot — Age by Sex (Hue: Survived, with Jitter)

```python
sns.stripplot(x='sex', y='age', data=df, jitter=True, hue='survived')
plt.title('Stripplot — Age by Sex (coloured by Survival)')
plt.show()
```

## 11. Stripplot — Age by Sex (with Jitter, no Hue)

```python
sns.stripplot(x='sex', y='age', data=df, jitter=True)
plt.title('Stripplot — Age by Sex')
plt.show()
```

## 12. Violinplot — Age by Sex (Hue: Survived)

```python
sns.violinplot(x='sex', y='age', data=df, hue='survived')
plt.title('Violinplot — Age by Sex and Survival')
plt.show()
```

## 13. Countplot — Passenger Count by Sex

```python
sns.countplot(x='sex', data=df)
plt.title('Passenger Count by Sex')
plt.show()
```
