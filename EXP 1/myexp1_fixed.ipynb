#### 4.b.2 Deal with Missing Data

**a) Drop data** — `dropna()` removes rows/columns with NaN values.
> Dropping is advised only when missing values are very few (0.01–0.5% of data).

```python
# Drops rows that have at least one NaN (original df unchanged)
df.dropna()
```

```python
# Original dataset remains unchanged
df
```

```python
# axis=1 → drop columns that have at least one NaN
df.dropna(axis=1)
```

```python
# Drop a row only if EVERY column in that row is NaN
df.dropna(how='all')
```

```python
# Keep only rows where "lunch" column has a value
# Note: lunch has 0 missing values in this dataset, so all 1000 rows are retained
df.dropna(subset=["lunch"])
```

**b) Replace Data** — Fill missing numeric values with the column mean using `fillna()`.

> **FIX:** All three score columns (`math score`, `reading score`, `writing score`) have missing values and must be filled — not just `reading score`.

```python
# Fill missing values in all three score columns with their respective column means
for col in ["math score", "reading score", "writing score"]:
    mean_val = df[col].mean()
    print(f"Mean of {col}: {mean_val:.2f}")
    df[col] = df[col].fillna(mean_val)

df
```

## 5. Data Formatting and Data Normalization
Check data types using `dtypes` and convert where needed using `astype()`.

```python
df.dtypes
```

> **FIX:** After filling NaN values, all score columns become `float64`. Since scores are whole numbers, convert them back to `int64`. Also convert all categorical text columns to `category` dtype.

```python
# Convert score columns from float64 back to int64 after fillna
df["math score"]    = df["math score"].astype("int64")
df["reading score"] = df["reading score"].astype("int64")
df["writing score"] = df["writing score"].astype("int64")

# Convert categorical text columns to category dtype
for col in ["gender", "race/ethnicity", "parental level of education",
            "lunch", "test preparation course"]:
    df[col] = df[col].astype("category")

df.dtypes
```

## 6. Turn Categorical Variables into Quantitative Variables

Two approaches:
- **Label Encoding** via `replace()` for binary columns (e.g., `gender`)
- **One-Hot Encoding** via `get_dummies()` for multi-class columns

### 6.a) Label Encoding for Binary Column (`gender`)

```python
# FIX: Result must be assigned back to df["gender"] — otherwise the mapping is discarded
df["gender"] = df["gender"].replace({"female": 0, "male": 1})
df["gender"]
```

### 6.b) One-Hot Encoding using `get_dummies()`

```python
# FIX: get_dummies() was described but never called — added here
# Converts multi-class categorical columns into separate 0/1 indicator columns
df_encoded = pd.get_dummies(
    df,
    columns=["race/ethnicity", "parental level of education",
             "lunch", "test preparation course"]
)
print("Shape after encoding:", df_encoded.shape)
df_encoded.head()
```
